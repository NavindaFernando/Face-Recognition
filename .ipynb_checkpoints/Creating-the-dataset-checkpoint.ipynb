```python
import os # os files handling
import numpy as np

# import dataset folder
data_path = r'train_data'

# the file names in the dataset are added to an list
labels = os.listdir(data_path)
print(labels)

# numbers are generated according to the number of files
categories = np.arange(len(labels))
print(categories)

# an dictionary is created using categories (file names) & labels (number of files)
categories_dict = dict(zip(labels,categories))
print(categories_dict)
```

```python
import cv2

data=[]
target=[]

# load the casecade classifier
face_classifier = cv2.CascadeClassifier('cascades/haarcascade_frontalface_default.xml')

# go to all files for one by one
for label in labels:
    
    # create path by useing data_path & label
    imgs_path = os.path.join(data_path,label)
    
    # image names in the created path are taken into a list.
    img_names = os.listdir(imgs_path)
    
    # go to all images for one by one
    for img_name in img_names:
        
        # get the paths of each image, because we want to load each images one by one
        img_path = os.path.join(imgs_path,img_name)
        img = cv2.imread(img_path) # load images in img_path
        
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # convert to gray image (BGR to GRAY)
        
        faces = face_classifier.detectMultiScale(gray) # get rectangle values (faces->2d array)
        
        # This loop runs according to the number of faces in the image
        for(x,y,w,h) in faces:
            
            #cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2) # drow a rectangle box
            cropped_face=gray[y:y+h,x:x+w] # crop faces 
            cv2.imshow('cropped_face',cropped_face) # show images (cropped faces) in window
            
            print('If you accept this cropped face press y or else n :')
            
            k=cv2.waitKey(0) # time to wait for images (holding)
            
            # check the y or n
            if(k==121):
                
                # if you press y, the image will be resized as a 50*50 image
                cropped_face = cv2.resize(cropped_face,(50,50))

                # if an image is accepted, add it to the data and target array.
                data.append(cropped_face)
                target.append(categories_dict[label])
            
cv2.destroyAllWindows()
```

```python
# convert numpy array
data = np.array(data)
target = np.array(target)

print(data.shape)
print(target.shape)

data = data.reshape(data.shape[0],data.shape[1]*data.shape[2]) # convert 1d (flattening)

# save physical array
np.save('data',data)
np.save('target',target)
```
